```python
from fastai.vision.all import *
```

```python
path = "data/colon_image_sets/"
```

```python
files = get_image_files(path)
```

```python
print(files[0])
print(str(files[0])[22:32])
print(str(files[0])[28] == 'a')
```

```python
def label_func(f): 
    return f[6] == 'a'
```

```python
dls = ImageDataLoaders.from_name_func(path, files, label_func, item_tfms=Resize(224))
```

```python
dls.show_batch()
```

```python
learn = vision_learner(dls, resnet34, metrics=error_rate)
```

```python
learn.fine_tune(1)
```

```python
learn.show_results()
```

```python
learn.lr_find()
```

```python
learn.fine_tune(3, 1e-3)
```

```python
learn.show_results()
```

```python
learn.export("models/modelo01_2022_12_05_08_40.pkl") # usar learn = load_learner() para carregar o modelo
```
